# Users_items

```python
df_items = pd.read_csv('users_items.csv')
```

```python
df_items.head()
```

```python
df_items = df_items.drop(columns="Unnamed: 0")
```

```python
#Parseo la columna items

import ast
def parse(items):
        return ast.literal_eval(items)
    
df_items['items'] = df_items['items'].apply(parse)
```

```python
df_items = df_items.drop(df_items[df_items["items_count"] == 0].index)
```

```python
df_items.shape
```

```python
df_items["items"][0]
```

```python
df_items.to_csv('users_items_clean.csv') 
```
