#### Endpoint 2

```python
#Traigo el Dataframe que contiene los precios de los juegos con sus respectivos Id para hacer matach con los 
#juegos que compro cada usuario. 

df_precios = pd.read_csv("preciosdejuegos.csv")
```

```python
df_precios.head()
```

```python
df_precios.rename(columns={'id': 'item_id'}, inplace=True)
```

```python
#Función para calcular el gasto de cada usuario
def calcular_gasto(usuario):
    gasto_total = 0
    for juego in usuario:
        item_id = juego['item_id']
        num = int(item_id)
        precio_fila = df_precios.loc[df_precios['item_id'] == num]  
        if not precio_fila.empty:
            precio = precio_fila["price"].values[0]
            gasto_total += precio
    return gasto_total

df['gasto_total'] = df['items'].apply(calcular_gasto)
```

```python
df_endpoint2 = df[['user_id', 'items_count', 'gasto_total']]
```

```python
#Guardo el df en un csv para ser consumido por la API.
df_endpoint2.to_csv("df_endpoint2.csv", index=False)
```

```python
df_games = pd.read_csv('cleaned_data.csv')
```

```python
small_games = df_games.head(5000)
small_items = df.head(50)
```

def UserForGenre( genero : str ): Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año de lanzamiento.
Ejemplo de retorno: {"Usuario con más horas jugadas para Género X" : us213ndjss09sdf, "Horas jugadas":[{Año: 2013, Horas: 203}, {Año: 2012, Horas: 100}, {Año: 2011, Horas: 23}]}

```python
def UserForGenre(genre):
    user_name = ''
    max_hours_count = 0
    for user in small_items.user_id:
        hours_count = 0
        for items in small_items.loc[small_items['user_id'] == user, 'items'].values[0]:
            game_id = items.get('item_id')
            if int(game_id) in df_games['id'].values:
                genres = df_games.loc[df_games['id'] == int(game_id)].values[0]
                if str(genre) in str(genres):
                    hours_count += int(items.get('playtime_forever'))

        if hours_count > max_hours_count:
            max_hours_count = hours_count
            user_name = user
    print("Usuario con más horas jugadas para Género {}: {} con {} horas.".format(genre, user_name, max_hours_count))
```
